## Exercise 1.5

```python
importlib.reload(F)
dict_sizes = range(20, 120, 1)  # Example range
errors = []
stds = []
retrievals = []


for M in dict_sizes:
    i=0
    mean_errors=[]
    std_errors=[]
    avgs_retrieved=[]
    while(i<15):
        mean_error, std_error, avg_retrieved = F.run_simulation_dictionary(M)
        i+=1
        mean_errors.append(mean_error)
        std_errors.append(std_error)
        avgs_retrieved.append(avg_retrieved)
    
    
    errors.append(np.mean(mean_errors))
    stds.append(np.mean(std_errors))
    retrievals.append(np.mean(avgs_retrieved))

# Plotting the results
plt.figure(figsize=(9.5,6))
plt.errorbar(dict_sizes, errors, yerr=stds, fmt='ro-', label='Error of pattern retrieval')
plt.xlabel('Dictionary Size (M)')
plt.ylabel('Error')
plt.legend()
plt.savefig("./Figures/error_of_retrieval_pattern.png")
plt.show()

plt.figure(figsize=(10,6))
plt.plot(dict_sizes, retrievals, 'ro-', label='Number of retrieved patterns')
plt.xlabel('Dictionary Size (M)')
plt.ylabel('Retrieved Patterns')
plt.legend()
plt.savefig("./Figures/number_of_retrieved_pattern.png")
plt.show()
```

The capacity is : $C=M_{max}/N=40/300$ 

## Exercise 1.6
Maximal number of patterns Mmax and retrieval errors beyond Mmax
 This is determined from the plot generated in Ex 1.5, where the number of retrieved patterns starts to decrease
significantly, and the error increases. This is usually where the network capacity C is reached.
 Mmax is the dictionary size M for which the number of retrieved patterns is maximized before the network's performance
 begins to degrade.
